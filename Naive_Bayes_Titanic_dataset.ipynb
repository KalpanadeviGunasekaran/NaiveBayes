```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("/content/titanic.csv")
df
```

```python
df.isnull().sum()
```

```python
fdf= df[['Survived', 'Pclass', 'Sex', 'Age', 'SibSp',
       'Parch', 'Cabin','Fare', 'Embarked']]
fdf.head()
```

```python
fdf["Sex"] = fdf["Sex"].replace(to_replace=fdf["Sex"].unique(), value = [1 , 0])
```

```python
#mean for normalised-losses
avg=fdf["Age"].astype("float").min(axis=0)
avg
```

```python
#Replace missing values with avg of age
fdf.loc[fdf["Age"].isna(), "Age"] = avg
```

```python
#mean for normalised-losses
avg=fdf["Fare"].astype("float").max(axis=0)
avg
```

```python
#Replace missing values with avg of age
fdf.loc[fdf["Fare"].isna(), "Fare"] = avg
```

```python
df["Cabin"].value_counts()
```

```python
df["Cabin"].replace(np.nan,'B96 B98',inplace=True)
```

```python
fdf.isnull().sum()
```

```python
fdf = pd.get_dummies(fdf, drop_first=True)
```

```python
fdf
```

```python
fdf.isnull().sum()
```

```python
train_y = fdf["Survived"]
train_x = fdf[['Pclass', 'Sex', 'Age', 'SibSp',
       'Parch', 'Fare', 'Embarked_Q','Embarked_S']]
```

```python
from sklearn.model_selection import train_test_split
train_data, val_data, train_target, val_target = train_test_split(train_x,train_y, train_size=0.8)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(train_data,train_target)
val_pred= model.predict(val_data)
```

```python
from sklearn.metrics import accuracy_score
print('Model accuracy score: {0:0.4f}'. format(accuracy_score(val_target, val_pred)*100)+ "%")
```
